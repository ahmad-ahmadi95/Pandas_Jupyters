üî∏ Our **Price** column is not **numeric**

üî∏ Its type is **object** or **string** here

```python
car_sales["Price"].dtype
```

```python
# Let's solve this problem
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '', regex=True).str[:-2]
```

```python
car_sales["Price"] = car_sales["Price"].astype(int)
```

```python
car_sales
```

```python
car_sales.Price.plot();
```

### 6. Manipulating Data with Pandas

```python
car_sales["Company"].str.lower()
```

```python
# Data with missing data
car_sales_missing = pd.read_csv("data/car-sales-missing-data.csv")
```

```python
car_sales_missing
```

#### üîπ Filling Missing Values in DataFrame

üî∏ `df["column_name"].fillna(value)`

```python
odometer_avg = round(car_sales_missing["Odometer"].mean())
odometer_avg
```

```python
car_sales_missing["Odometer"] = car_sales_missing["Odometer"].fillna(value=odometer_avg)
```

```python
car_sales_missing
```

#### üîπ Removing Missing Values in DataFrame

üî∏ `df.dropna()` 

```python
car_sales_missing.dropna(inplace=True)
```

```python
car_sales_missing
```

```python
car_sales_missing.to_csv("data/car-sales-missing-dropped.csv")
```

#### üîπ Creating Data From Existing Data

```python
seats = pd.Series([5, 5, 5, 5, 5])
```

```python
# Creating a new column
car_sales["Seats"] = seats
```

```python
car_sales
```

```python
# Syntax of new version of Pandas (future warning)
car_sales.fillna({"Seats": 5}, inplace=True)
```

```python
car_sales
```

```python
fuel_consumption = [7.5, 5, 6.8, 8.1, 9, 5.6, 6.9, 10, 5, 6.7]
car_sales["Fuel per 100km"] = fuel_consumption
```

```python
car_sales
```

#### üîπ Creating Data Using Column Arithmetic

```python
car_sales["Total fuel used"] = (car_sales["Odometer"] / 100) * car_sales["Fuel per 100km"]
```

```python
car_sales
```

#### üîπ Creating a Column From a Single Value

```python
car_sales["Number of wheels"] = 4
```

```python
car_sales
```

#### üîπ Removing Columns / Rows

üî∏ **For columns**: `df.drop(columns="column_name", axis=1)`

üî∏ **For rows**: `df.drop(labels="row_name", axis=0)`

```python
car_sales.drop(columns="Number of wheels", axis=1, inplace=True)
```

```python
car_sales
```

#### üîπ Shuffling DataFrame 

üî∏ **To mix up indices use:** `df.smaple(frac, random_state)`

```python
car_sales_shuffled = car_sales.sample(frac=1, random_state=7)
```

```python
car_sales_shuffled
```

```python
# Only select 30% of data
car_sales.sample(frac=0.3, random_state=7)
```

üî∏ As you see the **indices** has changed!

üî∏ Using `df.reset_index(drop=True)` we can **reset them to the default**!

```python
car_sales_shuffled.reset_index()
```

```python
car_sales.reset_index(drop=True)
```

#### üîπ Create New Data Using apply( ) Method

üî∏ `df["column_name"].apply(function)`: applies a function to a column

```python
# Convert Odometer (KM) -> Odometer (Mile)
car_sales["Odometer"] = car_sales["Odometer"].apply(lambda x:x*0.6)
```

```python
car_sales
```
